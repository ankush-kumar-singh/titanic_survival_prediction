# 1. Problem Statement

##### Objective:
 Predict whether a passenger survived the Titanic disaster using passenger information such as age, sex, class, fare, and family details.

# 2. Dataset Description

Dataset: Titanic Dataset (Kaggle)

Rows: 891
Target Variable: Survived

Features:
Pclass
Sex
Age
Fare
Embarked
SibSp
Parch
etc.

# 3. Data Cleaning

- Filled missing Age values using Pclass + Sex median.
- Filled missing Embarked values using mode.
- Dropped Cabin due to excessive missing values.
- Created HasCabin feature.
- Created FamilySize feature.
- Created IsAlone feature.

# 4. Exploratory Data Analysis

My findings:

- Females had significantly higher survival rates.
- First-class passengers survived more often.
- Higher fare passengers had better survival chances.
- Children survived more often than adults.
